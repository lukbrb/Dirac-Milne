# Spectre du bruit de fond cosmologique induit par les structures massives entre $z=0$ et $z=1080$ dans l'univers de Dirac-Milne

Le sujet consiste à estimer le *spectre du bruit de fond cosmologique* induit par les structures massives entre $z=0$ et la fenêtre de transparence de l'univers $z=1080$, dans ce que l'on appelle l'effet *Integrated Sachs-Wolfe* ([ISW](https://fr.wikipedia.org/wiki/Effet_Sachs-Wolfe)). Cet effet est prépondérant dans l'univers de Dirac-Milne, en raison de sa géométrie d'espace-temps plat et d'espace ouvert.

Pour l'estimation, on utilisera :
- Les caractéristiques géométriques de distances de l'univers de D-M
- La densité de matière de l'univers de D-M
- [Spectre de puissance](https://media4.obspm.fr/public/AAM/pages_ser-fourier/spec.html) des structures dans l'univers de D-M

*Ces trois éléments sont en pratique suffisant pour estimer avec une bonne approximation l'effet ISW et le spectre de perturbations induites sur le CMB.*

**Caractéristiques géométriques**

Pour les caracéristiques géométriques de l'univers de Dirac-Milne, il faut calculer :
- Le covolume entre *redshift* 0 et *redshift* z
- La distance radiale comobile en fonction du *redshift* z
- La distance de diamètre angulaire en fonction du *redshift* z

Combiné avec la masse des structures typiques à *redhsift* z, on peut **calculer le nombre moyen de structures en fonction du *redshift***. Chaque structure fait une *tâche* de modification sur la température du CMB sur le *ciel*. 

Finalement, on fait la somme de l'ensemble des structures pour chaque tranche de *redhsift*. (Pour chaque tranche de *redshift*, on fait la somme de toutes les structures, puis on fait la somme de toutes les tranches). Cette double somme permet d'obtenir le **spectre de modifications** de la température du fond diffus cosmologique.

# 1 - Calculs analytiques

## 1.1 - Distances dans l'univers de Dirac-Milne

D'après "[Introducing the Dirac-Milne Universe](https://arxiv.org/pdf/1110.3054.pdf)",  l'univers de Dirac-Milne n'est ni en accélération ni en décélération et est donc équivalent à un univers vide et ouvert. En terme des paramètres cosmologiques, cela correspond à $\Omega_m = 0$ et $\Omega_{\Lambda} = 0$, ou encore $\Omega_k = 1$. 
Il en découle que la distance de luminosité s'exprime de manière simple : 
$d_l(z) = \dfrac{c}{H_0}(1 + z) \sinh[ln(1 + z)]$. 

Et puisque l'on a $d_l(z) = (1 + z) d_m(z) = (1 + z)^2 d_a(z)$, on peut en déduire que : 
- la distance comobile radiale : $d_r(z) = \dfrac{c}{H_0}\ln(1 + z)$
- la distance comobile transversale : $d_m(z) = \dfrac{c}{H_0}\sinh[\ln(1 + z)]$
- la distance de diamètre angulaire : $d_a(z) = \dfrac{c}{H_0}\dfrac{\sinh[\ln(1 + z)]}{1 + z}$ 

De ces valeurs, on calcule ensuite le covolume et le volume propre:
- covolume : $V_c = 2\pi \left(\dfrac{c}{H_0}\right)^3 \times \left[f \sqrt{1 + f^2} - \ln (1 + z)\right]$ où $f=d_m(z) \dfrac{H_0}{c}$

- volume propre : $V_p = \frac{4}{3} \pi d_a(z)^3$

On trouve dans l'univers de Dirac-Milne $3,2$ protons par mètre cube. Connaissant la masse du proton $m_p = 1,6726 10^{-27} kg$, on peut en déduire que la densité de masse dans l'univers de Dirac-Milne $\rho_m = 5,35232 \times 10^{-27} kg.m^{-3}$. On peut ainsi déduire le paramètre de densité $\Omega_m = \frac{5,35232}{2,0}\times 10^{-1} = 0,267616$ de la matière. Pour l'antimatière on aurait l'opposé, et ainsi $\Omega_m = \Omega_{m+} + \Omega_{m-} = 0$. De plus, en partant de l'équation $2q = \Omega_m + \Omega_r - 2\Omega_{\Lambda}$ on a que $\Omega_m = 2\Omega_{\Lambda}$ puisque $\Omega_m = \Omega_r = 0$. 

## 1.2 - Distribution de structures dans l'univers de Dirac-Milne 

On s'intéresse tout d'abord à l'évolution de la densité en fonction du *redshift*. L'expansion de l'univers fait augmenter les volumes, et on a ainsi $\rho(z) = \rho_0 (1 + z)^3$. La densité décroît avec le temps. Connaître la densité de matière de l'univers à un *redshift* donné nous permet d'estimer la masse des structures à cette époque (en utilisant d'abord les données des $k_{pic}$), et donc leur nombre. La taille des structures est approximée par $L=\dfrac{\pi}{k_{pic}}$, et leur masse par $M = \frac{4}{3}\pi L^3 \times \rho(z)$. Le nombre de structures est ainsi $n(z) = \dfrac{\Delta V_c}{V_s(z)}$ où $V_s=\frac{4}{3}\pi L^3 $ est le volume d'une structure à $z$ donné. 

## 1.3 - Calcul de l'effet Sachs-Wolfe intégré

Dans la partie précédente nous avons estimé le nombre de structures, leurs masses et tailles en fonction du *redhsift*, dans un univers de Dirac-Milne. On a commencé en prenant une taille constante (à $z$ donné) pour les structures, correspondant au $k_{pic}$, telle que $L=\dfrac{\pi}{k_{pic}}$. De cette taille on a pu dériver la masse moyenne des structures à $z$ donné, et donc le nombre de structure. Par la suite nous tirerons la masse des structures selon une loi aléatoire en lognormal. 


Une fois le nombre de structures connu, on a *projeté* celles-ci sur le ciel, de manière aléatoire. Nous nous intéresserons dans cette partie à l'effet Sachs-Wolf intégré (ISW en anglais) que produisent les potentiels gravitationnels de ces structures (de matière). Pour ce faire, nous allons calculer le potentiel gravitationnel d'une structure, corrigé pour tenir compte de l'expansion, dans l'espace de Fourier, puis nous en déduiront l'expression de l'ISW, donnée par :

$\dfrac{\Delta T}{T} = \dfrac{2}{c^3} \int \dot{\Phi} a \mathrm{d}r$ 

Dans l'espace de Fourier, on a $\Phi_{+} = \dfrac{GM}{L} \times (1+z)$

# 3 - Précisions sur les fluctuations de température

## 3.1 - Redéfinition du potentiel gravitationnel

On a considéré précedemment des fluctuations créées en passant par des potentiels gravitationnels constants, laissant ainsi des tâches uniformes. Ce comportement n'est bien sûr pas réaliste et il est nécessaire de préciser la forme qu'a un potentiel, et comment ces derniers se répartissent dans l'espace.

```python
n_side = 64
pot = np.array(cc.CalculPot(n_side))
```

```python
pot -= pot[0][0] # On normalise le potentiel pour que le minimum soit à 0.
N = 5
M = np.zeros(( n_side * int(N), n_side * int(N)))
for i in range(1, N+1):
    for j in range(1, N+1):
        M[(i-1) * n_side : i * n_side, (j-1)*n_side : j * n_side] = pot
```

```python
fig, ax  = plt.subplots(figsize=(10, 10))
ax.imshow(M);
```

### Projection de `cosmocalc.Gpot.CalculPot()` sur l'ensemble $\left[-1, 1\right]$ :
- On veut une application linéaire et bijective
- On a donc $f(x) = ax + b$ et $f(0) = -1$ ainsi que $f(x_{max}) = 1$, soit $b=-1$ et $a=2/x_{max}$

On suppose ici que l'on a normalisé nos valeurs pour que $x_{min} = 0$.

```python
def normalise(tabx):
    tab = np.asarray(tabx)
    tab -= tab.min()
    a = 2/tab.max()
    return a * tab - 1
```

```python
plt.figure(figsize=(8,8))
plt.title("Potentiel gravitationnel normalisé")
plt.plot(normalise(pot))
plt.plot(np.ones(len(pot)), '--k')
plt.plot(-np.ones(len(pot)), '--k');
```

## 3.2 - Création de la nouvelle carte

```python
NSIDE = 2048
NPIX = hp.nside2npix(2048)
mat_test = np.zeros(NPIX)

for i in range(0, 140):
        # radius-ISW is the angular radius for a "SDSS" peak-size structure. We use it here as an approximation of the main structures
    radius_ISW = np.radians(theta_moy[i])
        # The parameter ampl_ISW takes into account the potential at the edge of the depletion zone
        # and the variation of this potential when the photon beam leaves it
    ampl_ISW = ISW[i]
    costheta_min = np.cos(radius_ISW)
    nb_clust_slice = n_z[i]
    

    for j in range(int(nb_clust_slice)):
    # Draw random unit vector for each cluster, defined by phi_clust and theta_clust, uniformly on sphere
        vec = cc.sample_sphere(1).T[0]
        # Get the pixels in the disk of the depletion zone around the direction of the center of the cluster
        ipix_disc = hp.query_disc(nside=NSIDE, vec=vec, radius=radius_ISW)
        vect_pix = hp.pix2vec(NSIDE, ipix_disc)
    
        costheta_pix = np.dot(vec, vect_pix)
        # Get the ISW weights for the pixels of this disk
        # Approximations used: spherical depletion zone and parallel photon beam crossing the structure
        ang_fact_ISW = 1./radius_ISW * np.sqrt(2*(costheta_pix - costheta_min))
        # Add this to the previous state of the map in ordrer to get the composite CMB image
        mat_test[ipix_disc] += ampl_ISW * ang_fact_ISW * normalise(ang_fact_ISW)
```

```python
mat_test -= mat_test.mean()  
hp.mollview(mat_test)
```

```python
LMAX = 2509
cl_from_map = hp.anafast(m, lmax=LMAX, use_pixel_weights=True)
ell = np.arange(len(cl_from_map))
cl_from_map *= ell * (ell +1) / (2*np.pi) * 1e12 # On passe de Cl[K^2] à l(l+1)Cl/2pi [µK^2]
```

```python
plt.figure(figsize=(10, 7))
plt.title("Spectre de puissance du FDC ", fontsize=18)
plt.plot(ell, cl_from_map, '.')
plt.xlabel('$\ell$', fontsize=17)
plt.ylabel('$ \dfrac{\ell(\ell + 1)}{2 \pi}C_\ell~[K^2]$', fontsize=15)
plt.grid()
```
